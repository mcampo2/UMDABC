```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")

school_data_complete.head()
```

# District Summary

```python
# Create a high level snapshot (in table form) of the district's key metrics


# school_data_complete["pass_math"] = student_data["math_score"]>=70
# school_data_complete["pass_reading"] = student_data["reading_score"]>=70
# school_data_complete["per_student_budget"]= school_data_complete["budget"]/school_data_complete["size"]

# calculate "Total Schools"
total_schools = school_data.shape[0]
# calculate "Total Students"
total_students = student_data.shape[0]
# calculate "Total Budget"
total_budget = school_data["budget"].sum()
# calculate "Average Math Score"
average_math_score = student_data["math_score"].mean()
# calculate "Average Reading Score"
average_reading_score = student_data["reading_score"].mean()
# calculate ""% Passing Math"
percent_passing_math = (school_data_complete["math_score"]>=70).sum()/student_data.shape[0]*100
# calculate ""% Passing Reading"
percent_passing_reading = (school_data_complete["reading_score"]>=70).sum()/student_data.shape[0]*100
# calculate "Overall Passing Rate" (Average of the above two)
overall_passing_rate = (average_math_score+average_reading_score)/2

district_key_metrics = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing Rate": [overall_passing_rate]
})

district_key_metrics
```

# School Summary


```python
# Create an overview table that summarizes key metrics about each school

# "School Name" name is an index and does not need to be calculated
# calculate "School Type"
school_type = school_data[["school_name", "type"]].set_index("school_name")["type"]
# calculate "Total Students"
total_students = school_data_complete[["school_name", "Student ID"]].groupby("school_name").count()["Student ID"]
# calculate "Total School Budget"
total_school_budget = school_data[["school_name", "budget"]].set_index("school_name")["budget"]
# calculate "Per Student Budget"
per_student_budget = total_school_budget / total_students
# calculate "Average Math Score"
average_math_score = school_data_complete[["school_name", "math_score"]].groupby("school_name").mean()["math_score"]
# calculate "Average Reading Score"
average_reading_score = school_data_complete[["school_name", "reading_score"]].groupby("school_name").mean()["reading_score"]
# calculate "% Passing Math"
percent_passing_math = school_data_complete[school_data_complete["math_score"]>=70][["school_name", "math_score"]] \
                       .groupby("school_name").count()["math_score"] / total_students * 100
# calculate "% Passing Reading"
percent_passing_reading = school_data_complete[school_data_complete["reading_score"]>=70][["school_name", "reading_score"]] \
                       .groupby("school_name").count()["reading_score"] / total_students * 100
# calculate % "Overall Passing Rate" (Average of the above two)
percent_overall_passing_rate = (percent_passing_math + percent_passing_reading)/2

# Create a dataframe to hold the above results
school_key_metrics = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_students,
    "Total School Budget": total_school_budget.map("${:,.2f}".format),
    "Per Student Budget": per_student_budget.map("${:,.2f}".format),
    "Average Math Score": average_math_score,#.map("{:,.2f}".format),
    "Average Reading Score": average_reading_score,#.map("{:,.2f}".format),
    "% Passing Math": percent_passing_math,#.map("{:,.2f}%".format),
    "% Passing Reading": percent_passing_reading,#.map("{:,.2f}%".format),
    "% Overall Passing Rate": percent_overall_passing_rate,#.map("{:,.2f}%".format)
})

school_key_metrics
```

# Scores by School Spending

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student).

# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# Use 4 reasonable bins to group school spending.
school_data_complete["spending"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_bins, labels=group_names)

# calculate "Average Math Score"
average_math_score = school_data_complete[["spending", "math_score"]].groupby("spending").mean()["math_score"]
# calculate "Average Reading Score"
average_reading_score = school_data_complete[["spending", "reading_score"]].groupby("spending").mean()["reading_score"]
# get total count of student by spending group, this will then be used to calculate later percents
count_total = school_data_complete[["spending", "Student ID"]].groupby("spending").count()["Student ID"]
# calculate "% Passing Math"
count_passing_math = school_data_complete[school_data_complete["math_score"]>=70][["spending", "math_score"]].groupby("spending").count()["math_score"]
percent_passing_math = count_passing_math / count_total * 100
# calculate "% Passing Reading"
count_passing_reading = school_data_complete[school_data_complete["reading_score"]>=70][["spending", "reading_score"]].groupby("spending").count()["reading_score"]
percent_passing_reading = count_passing_reading / count_total * 100
# calculate % "Overall Passing Rate" (Average of the above two)
percent_overall_passing_rate = (percent_passing_math + percent_passing_reading) / 2

# Create a dataframe to hold the above results
scores_by_spending = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing Rate": percent_overall_passing_rate
})

scores_by_spending.index.name = "Spending Ranges (Per Student)"
scores_by_spending
```

# Scores by School Size

```python
# Perform the same operations as above, based on school size.

# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Use 3 reasonable bins to group school sizes.
school_data_complete["size_bin"] = pd.cut(school_data_complete["size"], size_bins, labels=group_names)

# calculate "Average Math Score"
average_math_score = school_data_complete[["size_bin", "math_score"]].groupby("size_bin").mean()["math_score"]
# calculate "Average Reading Score"
average_reading_score = school_data_complete[["size_bin", "reading_score"]].groupby("size_bin").mean()["reading_score"]
# get total count of student by size group, this will then be used to calculate later percents
count_total = school_data_complete[["size_bin", "Student ID"]].groupby("size_bin").count()["Student ID"]
# calculate "% Passing Math"
count_passing_math = school_data_complete[school_data_complete["math_score"]>=70][["size_bin", "math_score"]].groupby("size_bin").count()["math_score"]
percent_passing_math = count_passing_math / count_total * 100
# calculate "% Passing Reading"
count_passing_reading = school_data_complete[school_data_complete["reading_score"]>=70][["size_bin", "reading_score"]].groupby("size_bin").count()["reading_score"]
percent_passing_reading = count_passing_reading / count_total * 100
# calculate % "Overall Passing Rate" (Average of the above two)
percent_overall_passing_rate = (percent_passing_math + percent_passing_reading) / 2

# Create a dataframe to hold the above results
scores_by_school_size = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing Rate": percent_overall_passing_rate
})

scores_by_school_size.index.name="School Size"
scores_by_school_size
```
